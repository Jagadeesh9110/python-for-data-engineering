# Class Variables vs Instance Variables

Since many developers come to Python from Java, this topic can be confusing.

Think of it like this:

| Java | Python |
|------|--------|
| Instance Field | Instance Variable (`self.variable`) |
| Static Field | Class Variable |

The most important difference is **where the data is stored**.

---

## 1. Instance Variables

Instance variables belong to an **individual object**.

They are created inside the constructor using `self`.

Every object gets its own copy.

```python
class Student:

    def __init__(self, name):
        self.name = name
```

Memory

```
std1
------
name = "Jag"

std2
------
name = "Krishna"
```

Changing one object's instance variable does **not** affect another object.

---

## 2. Class Variables

Class variables belong to the **class itself**, not to individual objects.

They are declared directly inside the class, outside any method.

```python
class Student:

    course = "Python"
```

Memory

```
Student Class
------------------
course = "Python"
------------------
```

Only **one copy** of a class variable exists.

Every object shares that single copy.

---

## Attribute Lookup

When Python executes

```python
print(std1.course)
```

it follows this order:

```
1. Look inside std1

        ↓

2. If not found, look inside Student class

        ↓

3. If found, return it
```

This process is called **attribute lookup**.

---

## What happens when we write

```python
std1.course = "ML"
```

Many beginners think this changes the class variable.

It does **not**.

Python creates a **new instance variable** named `course` inside `std1`.

Memory becomes

```
Student Class
------------------
course = "Python"
------------------

std1
------------------
name = "Jag"
course = "ML"
------------------

std2
------------------
name = "Krishna"
------------------
```

Now

```python
print(std1.course)
```

returns

```
ML
```

because `std1` now has its own `course`.

But

```python
print(std2.course)
```

returns

```
Python
```

because `std2` does not have a `course`, so Python reads it from the class.

---

## To change the shared class variable

Use the class name.

```python
Student.course = "Machine Learning"
```

Now every object that does **not** have its own `course` attribute will see the updated value.

---

## Summary

### Instance Variable

- Declared using `self.variable`
- Stored inside each object
- Every object has its own copy
- Changes affect only that object

### Class Variable

- Declared directly inside the class
- Stored inside the class
- Only one shared copy exists
- Accessed by all objects unless an object creates its own variable with the same name

---

```java

}
```

# Encapsulation

Encapsulation is one of the four fundamental principles of Object-Oriented Programming (OOP).

The word **Encapsulation** means **"wrapping or bundling things together."**

In OOP, encapsulation is the process of:

- Bundling data (variables) and the methods that operate on that data into a single unit (class).
- Controlling how the data can be accessed or modified.

A class is the basic unit of encapsulation.

---

# Name Mangling and `__dict__`

When Python sees a variable that starts with **double underscores (`__`)**, it does **not** store the variable using its original name.

For example,

```python
self.__name = name
```

is **not** stored internally as

```python
__name
```

Instead, Python automatically renames it to

```python
_Student__name
```

where

- `Student` → Class Name
- `__name` → Original Variable Name

This automatic renaming process is called **Name Mangling**.

---

## Why does Python perform Name Mangling?

Python does **not** have true private variables like Java.

Instead, it changes the variable's name internally to

- Reduce accidental access from outside the class.
- Prevent name collisions when inheritance is used.

Remember,

Name Mangling **does not make a variable truly private**.

It simply changes its internal name.

---

## What is `__dict__`?

Every normal Python object contains a special attribute called

```python
__dict__
```

`__dict__` is a dictionary that stores **all the instance variables** of that object.

Think of it as the object's **internal storage**.

Whenever you create an instance variable using

```python
self.variable = value
```

Python stores it inside the object's `__dict__`.

It is very useful for debugging because it lets you see exactly what attributes are stored inside an object.

````python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

std = Student("Jag", 22)

print(std.__dict__) # Python stores the instance variables exactly as they were created.
````

````python
class Student:

    def __init__(self, name):
        self.__name = name

std = Student("Jag")

print(std.__dict__) # Name Mangling- Although we wrote,Python did not store as __name Instead, Python automatically renamed it to _Student__name
## so we will get the AttributeError: 'Student' object has no attribute '__name'
````

# Inheritance

Inheritance is one of the four fundamental principles of Object-Oriented Programming (OOP).

Inheritance is the mechanism by which one class acquires the properties and methods of another class.

It allows us to reuse existing code instead of writing the same code multiple times.

---

## Why do we need Inheritance?

Suppose we have two classes:

- Student
- Teacher

Both have common information such as

- Name
- Age
- Address

Without inheritance, we would write the same variables and methods in both classes.

Inheritance allows us to place the common functionality in one class and reuse it in other classes.

This improves:

- Code Reusability
- Maintainability
- Extensibility
- Readability

## Without Inheritance

- Notice that both classes contain exactly the same code.

- This violates the DRY (Don't Repeat Yourself) principle.

````python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name)
        print(self.age)


class Teacher:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name)
        print(self.age)
````

## With Inheritance

````python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name)
        print(self.age)


class Student(Person):
    pass


class Teacher(Person):
    pass
````

# Method Overriding

Method Overriding occurs when a child class provides its own implementation of a method that already exists in the parent class.

The child class method replaces the parent class method for objects of the child class.

---

## Why do we need Method Overriding?

Sometimes the parent class provides a general implementation.

Different child classes may require different behavior.

Method overriding allows each child class to provide its own implementation while keeping the same method name.

---

## Java vs Python

Java

```java
class Animal{
    void sound(){}
}

class Dog extends Animal{

    @Override
    void sound(){}
}
```

Python

```python
class Animal:

    def sound(self):
        ...

class Dog(Animal):

    def sound(self):
        ...
```

Unlike Java, Python does not require an `@Override` annotation.

````python
class Animal:

    def sound(self):
        print("Animals make different sounds.")


class Dog(Animal):

    def sound(self):
        print("Dog barks.")

dog = Dog()

dog.sound()
````

- Although Dog inherited sound() from Animal, it defines its own version.

- Python always gives priority to the method defined in the child class.

````python
class Employee:

    def work(self):
        print("Employee is working.")


class Developer(Employee):

    def work(self):
        print("Developer is writing code.")


class Tester(Employee):

    def work(self):
        print("Tester is testing software.")

dev = Developer()

tester = Tester()

dev.work()

tester.work()
````

##  If the Child Does Not Override

````python
class Animal:

    def sound(self):
        print("Animals make different sounds.")


class Dog(Animal):
    pass


dog = Dog()

dog.sound()
````

# The `super()` Function

`super()` is a built-in Python function used to access the methods and constructors of the parent class.

It allows a child class to reuse the functionality of its parent class instead of rewriting the same code.

---

## Why do we need `super()`?

Suppose a child class wants to add some extra functionality, but it also wants the parent class to perform its initialization.

Instead of copying the parent's code, the child class can simply call the parent using

```python
super()
```

This promotes

- Code Reusability
- Maintainability
- Cleaner Code

## Without super()

````python
class Person:

    def __init__(self, name):
        self.name = name


class Student(Person):

    def __init__(self, name, roll_no):
        self.name = name  # already exists in the parent class.We are rewriting the same code.This violates the DRY (Don't Repeat Yourself) principle.
        self.roll_no = roll_no


std = Student("Jag", 101)

print(std.name)
print(std.roll_no)
````

## Using super()

````python
class Person:

    def __init__(self, name):
        self.name = name


class Student(Person):

    def __init__(self, name, roll_no):

        super().__init__(name)

        self.roll_no = roll_no


std = Student("Jag", 101)

print(std.name)
print(std.roll_no)
````

# What does `super().__init__(name)` actually do?

Many beginners think that

```python
super().__init__(name)
```

creates a new object of the parent class.

**This is NOT true.**

Python does **not** create a separate `Person` object.

There is only **one object**, which is the **Student object**.

The parent constructor simply initializes the inherited part of that same object.

---

## Conceptual Working

When Python encounters

```python
super().__init__(name)
```

it behaves conceptually like

```python
Person.__init__(self, name)
```

Notice carefully:

```python
Person.__init__(self, name)
```

The `self` being passed is **the Student object**, **not** a Person object.

This means the parent constructor is initializing the Student object itself.

---

# Internal Working

Suppose we write

```python
std = Student("Jag", 101)
```

### Step 1: Python Creates the Object

Python first creates **one Student object**.

Memory:

```text
Student Object

{}
```

---

### Step 2: Python Calls the Child Constructor

Python executes

```python
Student.__init__(std, "Jag", 101)
```

where `std` is the newly created Student object.

---

### Step 3: `super()` Calls the Parent Constructor

Inside the child constructor, we write

```python
super().__init__(name)
```

Conceptually, Python treats this as

```python
Person.__init__(std, "Jag")
```

Notice that the object is still

```text
std
```

No new `Person` object is created.

The parent constructor simply initializes its own attributes on the existing Student object.

---

### Step 4: Parent Constructor Finishes

After

```python
Person.__init__(std, "Jag")
```

the object becomes

```text
Student Object

name = "Jag"
```

The parent has finished initializing the inherited attribute.

---

### Step 5: Control Returns to the Child Constructor

Execution returns to

```python
Student.__init__()
```

The remaining statements continue executing.

```python
self.roll_no = roll_no
```

The object now becomes

```text
Student Object

name = "Jag"
roll_no = 101
```

---

## Final Memory Representation

```text
Student("Jag", 101)

        │
        ▼

Create Student Object

{}

        │
        ▼

Student.__init__(std, "Jag", 101)

        │
        ▼

super().__init__(name)

        │
        ▼

Person.__init__(std, "Jag")

        │
        ▼

Student Object

name = "Jag"

        │
        ▼

Return to Student.__init__()

        │
        ▼

self.roll_no = 101

        │
        ▼

Final Student Object

name = "Jag"
roll_no = 101
```

---

## Important Points

- `super()` **does not create a new parent object**.
- Only **one object** exists throughout the entire process.
- The `self` passed to the parent constructor is the **Student object**.
- The parent constructor initializes the inherited attributes of the Student object.
- After the parent constructor completes, execution returns to the child constructor, which initializes the child-specific attributes.

## calling parent methods

````python
class Person:

    def display(self):
        print("Person Display")


class Student(Person):

    def display(self):

        super().display()

        print("Student Display")


std = Student()

std.display()
````

# Types of Inheritance

Inheritance can be classified into different types based on how classes are related.

Python supports the following types of inheritance:

1. Single Inheritance
2. Multilevel Inheritance
3. Hierarchical Inheritance
4. Multiple Inheritance
5. Hybrid Inheritance (Combination of the above)

Unlike Java, Python fully supports **Multiple Inheritance**.

Because Python allows multiple inheritance, it also needs a mechanism to decide which parent class should be searched first when the same method exists in multiple parent classes.

Python solves this problem using the **Method Resolution Order (MRO)**.

Before learning MRO, we will first understand each type of inheritance.

# 1. Single Inheritance

Single Inheritance means a child class inherits from exactly one parent class.

This is the simplest and most common form of inheritance.

Java and Python both support Single Inheritance.

````python
class Person:

    def display(self):
        print("Person")


class Student(Person):
    pass


std = Student()

std.display()
````

# 2. Multilevel Inheritance

In Multilevel Inheritance, one class inherits from another child class.

Each child becomes the parent of the next class.

Both Java and Python support Multilevel Inheritance.

````python
class LivingBeing:

    def breathe(self):
        print("Breathing")


class Person(LivingBeing):
    pass


class Student(Person):
    pass


std = Student()

std.breathe()
````

# 3. Hierarchical Inheritance

In Hierarchical Inheritance, multiple child classes inherit from the same parent class.

The parent class provides common functionality that is shared by all child classes.

Both Java and Python support Hierarchical Inheritance.

````python
class Person:

    def display(self):
        print("Person")


class Student(Person):
    pass


class Teacher(Person):
    pass
````

# 4. Multiple Inheritance

Multiple Inheritance means one child class inherits from more than one parent class.

Unlike Java, Python fully supports Multiple Inheritance.

This allows a child class to combine the functionality of multiple parent classes.

````python
class Father:

    def money(self):
        print("Father's Money")


class Mother:

    def care(self):
        print("Mother's Care")


class Child(Father, Mother):
    pass


c = Child()

c.money()
c.care()
````

# Does Multiple Inheritance Create a Problem?

Suppose both parent classes contain a method with the same name.

Now the child class inherits both methods.

Which method should be executed?

This creates an ambiguity known as the **Diamond Problem**.

````python
class Father:

    def show(self):
        print("Father")


class Mother:

    def show(self):
        print("Mother")


class Child(Father, Mother):
    pass


child = Child()

child.show()
````

Whenever a method or attribute is accessed, Python follows a predefined search order.

If the method is found, Python immediately stops searching.

If the method is not found, Python moves to the next class in the MRO.

# A New Question

So far, the MRO examples have looked simple.

You might think Python simply searches from left to right.

However, in complex inheritance hierarchies, a simple left-to-right search can produce inconsistent results.

To guarantee a consistent search order, Python uses a formal algorithm called **C3 Linearization**.

C3 Linearization generates a valid Method Resolution Order (MRO) for every class while preserving inheritance relationships and avoiding contradictions.

In the next section, we will learn how Python computes the MRO using the C3 Linearization algorithm.

# C3 Linearization

Method Resolution Order (MRO) is not created randomly.

Python computes the MRO using an algorithm called **C3 Linearization**.

The purpose of the C3 algorithm is to produce a single, consistent order for searching classes.

This order must satisfy three important rules:

1. A child class is searched before its parent classes.
2. The order of parent classes specified in the class definition is preserved.
3. A class appears only once in the MRO.

The C3 algorithm guarantees that these rules are always satisfied, even in complex multiple inheritance hierarchies.

````python
class A:
    pass


class B(A):
    pass


class C(A):
    pass


class D(B, C):
    pass


print(D.mro())
````

# Rules Followed by C3 Linearization

Python constructs the MRO while following these rules.

## Rule 1

The child class always comes before its parent classes.

Example

D

↓

B

↓

C

↓

A

Rule 1 is satisfied because D is searched before all its parents.

---

## Rule 2

The order of parent classes is preserved.

If we write

class D(B, C)

then

B must appear before C.

If we write

class D(C, B)

then

C must appear before B.

Python never changes the order specified by the programmer.

---

## Rule 3

A class appears only once.

Notice that both B and C inherit from A.

Even though A is reachable through two paths,

Python includes A only once in the final MRO.

# How Does Python Build the MRO?

Conceptually, Python performs the following steps.

Step 1

Start with the child class.

↓

Step 2

Look at the parent classes.

↓

Step 3

Merge the MROs of all parent classes.

↓

Step 4

Ensure

- parent order is preserved,
- no class appears twice,
- every child appears before its parents.

↓

Step 5

Produce one final search order.

This final order is called the Method Resolution Order (MRO).

# Runtime Polymorphism

Runtime Polymorphism occurs when a child class overrides a method of its parent class.

At runtime,

Python determines which implementation should execute based on the object's actual type.

This is achieved through **Method Overriding**.

````python
class Animal:

    def sound(self):
        print("Animal")


class Dog(Animal):

    def sound(self):
        print("Bark")


class Cat(Animal):

    def sound(self):
        print("Meow")


animals = [Dog(), Cat()]

for animal in animals:
    animal.sound() # The object itself decides which implementation should execute. Exactly like Java's runtime polymorphism.
````

````python
class Animal:

    def sound(self):
        print("Animal Sound")


class Dog(Animal):

    def sound(self):
        print("Bark")


class Cat(Animal):

    def sound(self):
        print("Meow")


animals = [
    Dog(),
    Cat(),
    Animal()
]

for animal in animals:
    animal.sound()
````

# Duck Typing

Duck Typing is one of the most important concepts in Python.

It is a form of polymorphism where Python focuses on **what an object can do**, rather than **what type the object is**.

In other words,

Python does not ask:

> "What class does this object belong to?"

Instead, Python asks:

> "Does this object have the required method or attribute?"

If the required method exists, Python executes it.

If not, Python raises an error.

This philosophy is called **Duck Typing**.

The name comes from the famous saying:

> "If it walks like a duck and quacks like a duck, then treat it as a duck."

In Python,

if an object behaves like the required object, Python accepts it, regardless of its class.

# Java Approach

Java primarily follows **type-based polymorphism**.

When a method expects an object of a particular type, the object passed must be compatible with that type.

The compatibility is established through:

- Inheritance
- Interface implementation

Java verifies this relationship at **compile time**.

If an object does not satisfy the declared type, the program fails to compile.

# Python Approach

Python follows a different philosophy.

Instead of checking the object's type,

Python checks whether the object provides the required behavior.

If the required method exists,

Python executes it.

Otherwise,


Python does **not** ask:

> "Is this object an Animal?"

Instead, Python asks:

> "Does this object have a `sound()` method?"

Since both objects implement `sound()`, Python executes the method successfully.

This programming style is called **Duck Typing**.

# Method Overloading

- Method Overloading means defining multiple methods with the **same name** but **different parameter lists**.

The compiler decides which method to call based on the number or type of arguments passed.

Method Overloading is a form of **Compile-Time Polymorphism**.

Java supports Method Overloading.

Python does **not** support Method Overloading in the same way.

# Why Doesn't Python Support Method Overloading?

Unlike Java,

Python does not store multiple methods having the same name.

A class behaves like a namespace (or dictionary) where method names are keys.

When Python encounters another method with the same name,

the new method simply replaces the previous one.

Therefore,

only the **last definition** remains inside the class.

# Does This Mean Python Cannot Handle Different Numbers of Arguments?

No.

Python simply uses a different approach.

Instead of Method Overloading,

Python provides several powerful alternatives:

1. Default Arguments
2. Variable-Length Arguments (*args)
3. Keyword Variable Arguments (**kwargs)
4. functools.singledispatch (Advanced)

These approaches provide the same flexibility without creating multiple methods having the same name.

## Alternative 1 — Default Arguments

````python
class Calculator:

    def add(self, a, b, c=0):
        return a + b + c


cal = Calculator()

print(cal.add(2,3))

print(cal.add(2,3,4))
````

## Alternative 2 — *args

````python
class Calculator:

    def add(self, *numbers):
        return sum(numbers)


cal = Calculator()

print(cal.add(2,3))

print(cal.add(2,3,4))

print(cal.add(2,3,4,5,6))
````

# Abstraction

Abstraction is the process of exposing only the essential functionality of an object while hiding its implementation details.

In other words,

Abstraction focuses on **what an object does**, not **how it does it**.

For example,

When you drive a car, you know how to

- Start the engine
- Accelerate
- Brake

However,

you do not need to know how the engine, gearbox, or fuel injection system works internally.

The car exposes only the necessary operations while hiding the internal implementation.

This is called **Abstraction**.

# Why Do We Need Abstraction?

Abstraction provides several benefits.

- Hides implementation complexity.
- Exposes only the required functionality.
- Reduces coupling between classes.
- Makes code easier to maintain.
- Allows different implementations while keeping the same interface.

The user of a class only needs to know **what** the class can do.

The implementation can change without affecting the user.

## Java Code example

abstract class Animal{

    abstract void sound();

}

class Dog extends Animal{

    @Override
    void sound(){
        System.out.println("Bark");
    }

}


The abstract class

Animal

declares the method

sound()

but does not provide an implementation.

Every child class must provide its own implementation.

Java enforces this rule during compilation.

# Python Approach

Python provides abstraction through the **abc** module.

The abc module stands for

**Abstract Base Classes**.

It allows us to create abstract classes and abstract methods.

Python provides two important components.

- ABC
- @abstractmethod

## Creting an Abstract Class

````python
from abc import ABC, abstractmethod


class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass
````

Animal

inherits from

ABC

making it an Abstract Base Class.

The decorator

@abstractmethod

marks

sound()

as an abstract method.

Any concrete subclass must implement this method.

````python
from abc import ABC, abstractmethod


class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):

    def sound(self):
        print("Bark")


dog = Dog()

dog.sound()
````

## What Happens If We Don't Implement It?

- TypeError:Can't instantiate abstract class Dog with abstract method sound

When Python sees

@abstractmethod

it records that

sound()

must be implemented.

When an object is created,

Python checks

"Have all abstract methods been implemented?"

If the answer is

No

↓

TypeError

If the answer is

Yes

↓

Object creation succeeds.

````python
from abc import ABC, abstractmethod


class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


circle = Circle(5)

print(circle.area())
````

# Abstraction vs Encapsulation

Many beginners confuse these concepts.

Encapsulation

- Bundles data and methods together.
- Controls access to data.
- Focuses on protecting object state.

Abstraction

- Hides implementation details.
- Shows only the required functionality.
- Focuses on simplifying the interface.

In simple terms,

Encapsulation answers

"How can we protect the data?"

Abstraction answers

"What functionality should be exposed to the user?"
