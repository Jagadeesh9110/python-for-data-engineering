# Variables and Objects

One of the biggest differences between Java and Python is how variables work.

Many beginners think

```python
x = 10
```

means

```
x stores 10
```

This is **not completely true**.

Python variables store **references** to objects.

Conceptually

```
x
│
│
▼
+---------+
| Integer |
|    10   |
+---------+
```

The integer object stores the value.

The variable stores a reference to that object.

Everything in Python is an object.

````python
x = 10

print(id(x))

name = "Python"

print(id(name))

numbers = [10, 20, 30]

print(id(numbers))
````

# id()

Every object has an identity.

Python provides

```python
id(object)
```

to view it.

Think of it as the object's unique identity during its lifetime.

Example

```
numbers

↓

List Object

↓

ID = 140589123456
```

Two different objects have different identities.

Time Complexity

```
O(1)
```

````python
a = [10, 20, 30]

b = a

print(a)

print(b)

print(id(a))

print(id(b))
````

# Assignment

This is the first important rule.

When you write

```python
b = a
```

Python does **NOT** copy the list.

Instead

Both variables refer to the same object.

Conceptually

```
a
│
│
├──────────────┐
│              │
▼              │
[10,20,30] ◄───┘
▲
│
b
```

No new list is created.

````python
a = [10, 20, 30]

b = a

b.append(40)

print(a)

print(b)
````

# Immutable Objects

Integers are immutable.

When you write

```python
b = 200
```

Python does NOT modify

```
100
```

Instead

Python creates

```
200
```

and changes

```
b
```

to refer to the new object.

The original object remains unchanged.

````python
a = 100

b = a

print(id(a))

print(id(b))

b = 200

print(a)

print(b)

print(id(a))

print(id(b))
````

# Mutable Objects

Lists are mutable.

Python changes the existing object.

No new list is created.

Therefore

Every reference sees the updated list.

````python
a = [1, 2, 3]

b = a

b[0] = 100

print(a)

print(b)
````

# copy()

Creates a **new list object**.

Conceptually

```
numbers

↓

[10,20,30]

copy_numbers

↓

[10,20,30]
```

Two different objects.

Changing one list does not affect the other.

This is called a **Shallow Copy**.

````python
numbers = [10, 20, 30]

copy_numbers = numbers.copy()

print(numbers)

print(copy_numbers)

print(id(numbers))

print(id(copy_numbers))


numbers = [10, 20, 30]

copy_numbers = numbers.copy()

copy_numbers.append(40)

print(numbers)

print(copy_numbers)
````

````python
numbers = [
    [1, 2],
    [3, 4]
]

copy_numbers = numbers.copy()

copy_numbers[0].append(100)

print(numbers)

print(copy_numbers)
````

# Why?

copy()

copies only the outer list.

The inner lists are still shared.

Memory

Outer List

↓

New

Inner Lists

↓

Shared

This is called a

Shallow Copy.

# deepcopy()

Creates completely independent objects.

Outer List

↓

Copied

Inner Lists

↓

Copied

Everything becomes independent.

This is called

Deep Copy.

````python
import copy

numbers = [
    [1, 2],
    [3, 4]
]

deep = copy.deepcopy(numbers)

deep[0].append(100)

print(numbers)

print(deep)
````

````python
employee = {
    "name": "Jagadeeswar",
    "skills": ["Python", "SQL"]
}

employee_copy = employee.copy()

employee_copy["skills"].append("Pandas")

print(employee)

print(employee_copy)
````

# range()

`range()` generates a sequence of numbers.

It is commonly used in `for` loops.

Syntax

```python
range(stop)
range(start, stop)
range(start, stop, step)
```

Important

- start → Included
- stop → Excluded
- step → Increment/Decrement

`range()` does NOT create a list.

It returns a **range object**, making it memory efficient.

````python
print(range(5))
print(type(range(5)))

for i in range(5):
    print(i)
````

````python
# stop
for i in range(5):
    print(i)

print("-"*20)

# start, stop
for i in range(2, 8):
    print(i)

print("-"*20)

# start, stop, step
for i in range(2, 12, 2):
    print(i)

print("-"*20)

# reverse
for i in range(10, 0, -2):
    print(i)
````

# Internal Working

When you write

```python
range(1000000)
```

Python does **NOT** create one million integers.

Instead, it stores only:

- start
- stop
- step

When needed, Python calculates the next value.

This is called **lazy generation**.

Memory Usage

```
start = 0
stop = 1000000
step = 1
```

instead of storing

```
0
1
2
...
999999
```

This makes `range()` extremely memory efficient.

````python
r = range(1000000)

print(r)

print(r[0])

print(r[999999])
````

# enumerate()

One of the most useful Python functions.

Instead of manually maintaining an index,

```python
index = 0

for value in data:
```

Python provides

```python
enumerate()
```

It returns

```
(index, value)
```

for every iteration.

````python
languages = ["Python","Java","Go","Rust"]

for index, language in enumerate(languages):
    print(index, language)
````

````python
languages = ["Python","Java","Go"]

for index, language in enumerate(languages, start=1):
    print(index, language)
````

# zip()

Used when multiple iterables should be processed together.

Returns pairs of corresponding elements.

Syntax

```python
zip(iterable1, iterable2)
```

Stops when the shortest iterable ends.

````python
names = ["John","Rahul","Sai"]

marks = [90,85,92]

for name, mark in zip(names, marks):
    print(name, mark)
````

````python
names = ["John","Rahul","Sai"]

marks = [90,85,92]

cities = ["Delhi","Hyderabad","Bangalore"]

for name, mark, city in zip(names, marks, cities):
    print(name, mark, city)
````

# iter()

Every iterable can produce an iterator.

An iterator knows

- Current Position
- Next Element

Syntax

```python
iterator = iter(iterable)
```

Most of the time,

Python calls `iter()` automatically in a `for` loop.

Understanding it helps explain how loops work internally.

````python
numbers = [10,20,30]

iterator = iter(numbers)

print(iterator)

print(type(iterator))
````

````python
numbers = [10,20,30]

it = iter(numbers)

print(next(it))

print(next(it))

print(next(it))

# Uncomment

# print(next(it))
````
