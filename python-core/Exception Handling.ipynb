# Exception Handling

An **exception** is an event that occurs during the execution of a program and interrupts the normal flow of the program.

In simple terms,

An exception represents an error or unexpected situation that occurs while the program is running.

Examples:

- Dividing by zero
- Accessing an invalid list index
- Opening a file that does not exist
- Converting an invalid string to an integer

If exceptions are not handled, the program terminates immediately.

Exception Handling allows us to detect these situations and respond appropriately instead of crashing the program.

````python
print("Program Started")

num = 10 / 0

print("Program Finished")
````

-  Notice that "Program Finished" is never printed because the program terminates when the exception occurs.

An Exception is an **object** created by Python when an error occurs.

Every exception in Python is represented by an object belonging to an Exception class.

Examples:

- ZeroDivisionError
- ValueError
- TypeError
- IndexError
- FileNotFoundError

These are all classes that inherit from Python's built-in `Exception` class.

# try Statement

The `try` block contains code that might raise an exception.

Python executes the statements inside the `try` block normally.

If no exception occurs,

the program continues normally.

If an exception occurs,

Python immediately stops executing the remaining statements inside the `try` block and starts searching for a matching `except` block.

## Example

````python
print("Program Started")

try:
    num = 10 / 0
    print("Division Successful")
except ZeroDivisionError:
    print("Cannot divide by zero")

print("Program Finished")
````

Python executes the code in this order.

```
Program Started

↓

Enter try block

↓

10 / 0

↓

ZeroDivisionError occurs

↓

Stop executing remaining statements inside try

↓

Search matching except block

↓

Execute except block

↓

Continue with remaining program

↓

Program Finished
```

## No Exception Case

````python
print("Program Started")

try:
    num = 10 / 2
    print(num)
    print("Division Successful")
except ZeroDivisionError:
    print("Cannot divide by zero")

print("Program Finished")
````

When Python enters a `try` block, it prepares to handle exceptions.

If every statement executes successfully,

the `except` block is skipped.

If an exception occurs,

Python immediately leaves the `try` block.

It then compares the exception with each `except` block until it finds a match.

After executing the matching `except` block,

execution continues with the first statement after the entire `try-except` structure.

# Multiple except Blocks

A program can raise different types of exceptions.

Instead of handling every exception in the same way,

Python allows us to define multiple `except` blocks.

Each `except` block handles a specific type of exception.

Python executes **only the first matching `except` block**.

````python
try:
    num = 10 / 0

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid Value")
````

````python
try:
    age = int("Hello")

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid Integer")
````

Suppose the following code is executed.

```python
try:
    age = int("Hello")
except ZeroDivisionError:
    ...
except ValueError:
    ...
except TypeError:
    ...
```

Python performs the following steps.

```
Exception Raised

↓

ValueError Object Created

↓

Compare with ZeroDivisionError

↓

No Match

↓

Compare with ValueError

↓

Match Found

↓

Execute ValueError block

↓

Skip remaining except blocks

↓

Continue program
```

````python
try:
    int("Hello")

except ValueError:
    print("Handled ValueError")

except Exception:
    print("Handled Exception")
````

## Even though Exception can handle almost every exception, Python executes only the first matching except block.

````python
try:
    numbers = [10, 20]

    print(numbers[5])

except Exception:
    print("Some error occurred")
````

Always write exception handlers from

**Most Specific**

↓

**Least Specific**

Example

```python

except ValueError:

```

Never place `Exception` before more specific exception classes.

## Catch Multiple Exceptions Together

# else Block

The `else` block is optional.

It is executed **only if the `try` block completes successfully without raising any exception**.

If an exception occurs,

the `else` block is skipped.

The `else` block is useful for writing code that should run only when no errors have occurred.

## Syntax
```python
try:
    # Risky Code

except ExceptionType:
    # Handle Exception

else:
    # Executes only if no exception occurred
```


## Example - No Exception

````python
try:
    num = 10 / 2
    print("Division Successful")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("No Exception Occurred")
````

## Example - Exception Occurs

````python
try:
    num = 10 / 0
    print("Division Successful")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("No Exception Occurred")
````

```
Enter try

↓

Exception?

───────────────┐
               │
No             Yes
│              │
▼              ▼
Execute else   Execute except
│              │
└──────┬───────┘
       ▼
 Continue Program
```

````python
try:
    file = open("students.txt")

except FileNotFoundError:
    print("File not found")

else:
    print("File opened successfully")
    file.close()
````

- The code inside else runs only if the file was opened successfully. This avoids writing code that assumes the operation succeeded when it may have failed.

# Finally Block

The `finally` block is used for cleanup.

It executes **whether an exception occurs or not**.

Typical uses include:

- Closing files
- Closing database connections
- Releasing locks
- Closing network sockets
- Cleaning temporary resources

The `finally` block is designed to ensure cleanup code always runs.

```python
try:
    # Risky Code
    pass
except Exception as e:
    # Handle Exception
    pass
finally:
    # Cleanup Code
    pass
```


## Example - Exception Occurs

````python
try:
    print("Inside try")

    num = 10 / 0

except ZeroDivisionError:
    print("Handled Exception")

finally:
    print("Cleanup Done")
````

## Example - No Exception

````python
try:
    print("Inside try")

    num = 10 / 2

except ZeroDivisionError:
    print("Handled Exception")

finally:
    print("Cleanup Done")
````

- The finally block executes in both cases.

```

↓

│              │
▼              ▼

│              │
└──────┬───────┘
```

````python
file = None

try:
    file = open("students.txt", "r")
    data = file.read()

except FileNotFoundError:
    print("File not found")

finally:
    if file:
        file.close()
        print("File Closed")
````

## Complete Flow (try → except → else → finally)

```python

    print("Try")

    print("Except")

    print("Else")

    print("Finally")
```

### Output
```text

```

### How It Works

*   **try**: The code inside this block runs first. If it succeeds without errors, it moves to the `else` block.
*   **except**: This block is skipped because no exception occurred. It only runs if the `try` block encounters an error.
*   **else**: This block runs **only if** the `try` block executes successfully without raising any exceptions.
*   **finally**: This block **always runs** at the very end, regardless of whether an error occurred or not.


| Situation             | Execution Order                      |
| --------------------- | ------------------------------------ |
| No Exception          | `try → else → finally`               |
| Exception Handled     | `try → except → finally`             |
| Exception Not Handled | `try → finally → Program Terminates` |


# raise Statement

Until now, we have handled exceptions raised automatically by Python.

Examples:

- ZeroDivisionError
- ValueError
- IndexError

However,

there are situations where our program detects an invalid condition before Python does.

In such cases,

we can explicitly raise an exception using the `raise` statement.

The `raise` statement creates an exception object and interrupts the normal flow of the program.

## Custom Exceptions in Python

### Syntax
```python
raise ExceptionType("Error Message")
```

### Example
```python
raise ValueError("Invalid Age")
```

### How It Works

*   **raise**: This keyword is used to manually trigger a specific exception in Python.
*   **ExceptionType**: The specific built-in error class (like `ValueError`, `TypeError`, or `IndexError`) that matches the issue.
*   **"Error Message"**: The descriptive text string passed into the exception to explain exactly why the code failed.


````python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative") 
    #The exception was not raised by Python automatically.Our code detected invalid input and explicitly raised a ValueError.

print(age)
````

When Python executes

```python
raise ValueError("Invalid Age")
```

it performs the following steps.

```
Create ValueError Object

↓

Search for Matching except Block

↓

Found?

             │
Yes          No
│            │
▼            ▼
Execute      Terminate Program
except
```

## Raising Different Exceptions

````python
name = ""

if not name:
    raise ValueError("Name cannot be empty")
````

````python
age = "Twenty"

if not isinstance(age, int):
    raise TypeError("Age must be an integer")
````

````python
raise FileNotFoundError("Configuration file missing")
````

````python
try:
    age = -2

    if age < 0:
        raise ValueError("Negative age is not allowed")

except ValueError as e:
    print("Handled:", e)
````

## Why use raise?

The `raise` statement allows us to enforce business rules.

Examples:

- Age cannot be negative.
- Salary cannot be less than zero.
- Password must contain at least 8 characters.
- Email format is invalid.
- Bank balance cannot become negative.

Python cannot know these application-specific rules.

Therefore,

our program raises the appropriate exception when a rule is violated.

## Language Comparison: Raising Exceptions

### Java
```java

```

### Python
```python

```

### Key Differences

*   **Keywords**: Java uses the **`throw`** keyword along with **`new`** to instantiate the error object, while Python uses the **`raise`** keyword directly.
*   **Exception Classes**: Java uses **`IllegalArgumentException`** to signal that a method has been passed an illegal argument, whereas Python uses the built-in **`ValueError`**.
*   **Syntax**: Java requires a **semicolon** at the end of the line and parentheses for instantiation, while Python relies on its cleaner, semicolon-free syntax.

# Custom Exceptions

Python provides many built-in exceptions.

However,

real-world applications often need domain-specific exceptions.

Examples:

- InvalidEmployeeID
- InvalidTransaction
- InsufficientBalance
- InvalidOrder
- StudentNotFound

For these situations,

Python allows us to create our own exception classes.

## Creating a Custom Exception

````python
class InvalidAgeError(Exception): # InvalidAgeError inherits from Python's built-in Exception class.
    pass
````

````python
class InvalidAgeError(Exception):
    pass


age = -10

if age < 0:
    raise InvalidAgeError("Age cannot be negative")
````

## Handling a Custom Exception

````python
class InvalidAgeError(Exception):
    pass


try:
    age = -10

    if age < 0:
        raise InvalidAgeError("Age cannot be negative")

except InvalidAgeError as e:
    print(e)
````

## Custom Exception with Attributes

````python
class InvalidSalaryError(Exception):

    def __init__(self, salary):
        self.salary = salary
        super().__init__(f"Invalid Salary: {salary}")


try:
    salary = -5000

    if salary < 0:
        raise InvalidSalaryError(salary)

except InvalidSalaryError as e:
    print(e)
````

```
BaseException

       ▼
 Exception

       ├── TypeError
       ├── IndexError
       ├── FileNotFoundError
       ├── ZeroDivisionError
       │
       └── InvalidAgeError
             └── InvalidSalaryError
```

Your own exception classes should generally inherit from `Exception`, not directly from `BaseException`.

- Raise the most appropriate exception type.
- Use clear and descriptive error messages.
- Create custom exceptions for application-specific errors.
- Inherit custom exceptions from `Exception`.
- Avoid using generic `Exception` when a more specific exception better describes the problem.

# Exception Objects and the `as` Keyword

When an exception occurs,

Python creates an **Exception Object**.

This object contains useful information about the error, such as:

- Error message
- Exception type
- Additional attributes (for some exceptions)

We can access this object using the `as` keyword.

## Syntax
```python
try:
    # Risky Code

except ExceptionType as e:
    # e is the exception object
```

````python
try:
    num = int("Hello")

except ValueError as e:
    print(e)
````

`e` is an object of type `ValueError`.

When we execute:

```python
print(e)
```

Python calls:

```python
print(str(e))
```

which prints the exception message.

## Accessing the Exception Type

````python
try:
    10 / 0

except Exception as e:
    print(type(e))
````

## Accessing the Error Message
- Most exception objects store their constructor arguments in the args attribute.

````python
try:
    int("ABC")

except Exception as e:
    print(e.args)
````

## Multiple Information

````python
try:
    numbers = [1, 2]

    print(numbers[5])

except Exception as e:
    print("Type :", type(e).__name__)
    print("Message :", e)
    print("Arguments :", e.args)
````

# Re-raising Exceptions

Sometimes we catch an exception,

perform some work (such as logging),

and then allow the exception to continue propagating.

This is called **Re-raising** an exception.

Python allows this using:

```python
raise
```

without specifying an exception object.

````python
try:
    10 / 0

except ZeroDivisionError:
    print("Logging the error...")

    raise
````

Python performs the following steps.

```
Exception Occurs

↓

Enter except Block

↓

Perform Logging

↓

raise

↓

Raise the SAME Exception Again

↓

Search for Another Handler

↓

If none found

↓

Terminate Program
```

Re-raising is useful when:

- Logging an error
- Cleaning temporary resources
- Updating metrics
- Performing auditing

After these actions,

the original exception is allowed to propagate so that higher-level code can handle it.

## Exception Chaining

- Sometimes, while handling one exception,another exception occurs.

- Python preserves both exceptions.

````python
try:
    int("ABC")

except ValueError as e:
    raise RuntimeError("Processing Failed")
````

## Explicit Exception Chaining
- Python also allows us to explicitly indicate the cause.

````python
try:
    int("ABC")

except ValueError as e:
    raise RuntimeError("Invalid User Input") from e
````

Using:

```python
raise NewException(...) from old_exception
```

creates a clear relationship between the original exception and the new one.

This is especially useful in large applications where low-level exceptions are converted into higher-level business exceptions while preserving the original cause.

- Catch only exceptions you can handle.
- Preserve the original exception when appropriate using `raise` or `raise ... from ...`.
- Use `as e` to access exception details.
- Log exceptions before re-raising them in production applications.
- Avoid hiding exceptions unless recovery is intentional and safe.
