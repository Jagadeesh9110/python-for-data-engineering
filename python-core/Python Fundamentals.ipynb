# Strings in Python

A string is an **ordered, immutable sequence of Unicode characters**.

Unlike Java:

- Python has **no separate `char` data type**.
- Even a single character is a string.

Examples

```python
name = "Jagadeeswar"

city = 'Kadapa'

letter = "A"
```

```python
print(type(name))
print(type(letter))
```

Output

```python
<class 'str'>
<class 'str'>
```

Notice that `"A"` is **not** a character.

It is a string of length 1.

### Characteristics

- Ordered
- Immutable
- Indexed
- Supports duplicate characters
- Supports Unicode

````python
# Using Double Quotes
name = "Jagadeeswar"

# Using Single Quotes
city = 'Kadapa'

# Triple Quotes (Multi-line Strings)
message = """
Welcome
To
Python
"""

# Raw String (Useful for Windows paths)
path = r"D:\Python\Projects"

print(name)
print(city)
print(message)
print(path)
````

# Internal Working

When you write

```python
name = "Python"
```

Python creates a **String Object**.

Conceptually

```
name
 │
 ▼
+----------------+
|   "Python"     |
+----------------+
```

The variable stores a **reference** to the string object.

The object internally stores:

- Characters
- Length
- Hash (cached when needed)
- Other implementation details

Remember:

**Variables do NOT store the actual string.**

They store a **reference** to the string object.

# String Immutability

Strings cannot be modified after they are created.

Python does **not** allow changing individual characters.

Attempting to do so raises a TypeError.

Why?

Because strings are immutable.

````python
language = "Python"

print(language)

# Uncomment to observe the error

# language[0] = "J"
````

# What Happens During Modification?

Example

```python
language = "Python"

language = "Jython"
```

Python does NOT modify the original string.

Instead:

Step 1

```
language
   │
   ▼
"Python"
```

Step 2

Python creates

```
"Jython"
```

Step 3

Now

```
language
   │
   ▼
"Jython"
```

The old string becomes eligible for garbage collection if nothing else refers to it.

This is the meaning of **immutability**.

# Why Python Makes Strings Immutable

## 1. Safety

If multiple variables refer to the same string,
one variable cannot accidentally modify it.

---

## 2. Performance

Python can reuse immutable strings internally.

---

## 3. Dictionary Keys

Strings are hashable.

Example

```python
employee = {
    "name": "Jagadeeswar"
}
```

If strings were mutable,

their hash values could change,

breaking dictionary lookups.

This is one of the major reasons strings are immutable.

````python
first = "Hello"

second = "World"

# Concatenation
print(first + " " + second)

# Repetition
print(first * 3)

# Length
print(len(first))

# Membership
print("H" in first)

print("z" in first)
````

# String Slicing

Slicing extracts a portion of a string.

Syntax

```python
string[start:stop]
```

Important Rule

- start → Included
- stop → Excluded

Python returns a **new string**.

Original string is never modified.

````python
language = "Python"

print(language[0:3])
print(language[1:4])
print(language[2:6])
````

````python
language = "Python"

print(language[1:5])
````

````python
language = "Python"

print(language[:4])

print(language[2:])

print(language[:])
````

# Slice Step

Syntax

```python
string[start:stop:step]
```

Step decides how many positions Python moves.

Default step

```
1
```

````python
language = "PythonProgramming"

print(language[0:10:2])

print(language[::2])

print(language[1::2])
````

````python
language = "Python"

print(language[::-1]) #This creates a **new reversed string**.Original string remains unchanged.

print(language)
````

````python
text = "EnterprisePython"

print(text[:10])

print(text[10:])

print(text[5:12])

print(text[::3])

print(text[-5:])

print(text[:-5])

print(text[-8:-2])

print(text[2:-2])

print(text[::-2])
````

````python
text = "Python"

copy = text[:]

print(text)

print(copy)

print(text is copy)

print(text == copy)
````

# String Case Conversion

These methods create a **new string**.

| Method | Description |
|----------|-------------|

| upper() | Converts every character to uppercase |
| capitalize() | First character uppercase |
| title() | First letter of every word uppercase |

Time Complexity

# format()

A newer formatting approach.

Syntax

```python

```

````python
name = "Jagadeeswar"
company = "Boeing"

print("Welcome {}".format(name))

print("{} works at {}".format(name, company))

print("{1} works at {0}".format(company, name))

print("Employee : {name} Company : {company}".format(
    name=name,
    company=company
))
````

# f-Strings

Introduced in Python 3.6.

This is the preferred formatting style.

Syntax

```python
f"...{variable}..."
```

Advantages

- Fast
- Readable
- Supports expressions
- Preferred in enterprise applications

````python
name = "Jagadeeswar"
salary = 50000

print(f"Employee : {name}")

print(f"Salary : {salary}")

print(f"Updated Salary : {salary + 5000}")

print(f"Salary in Lakhs : {salary/100000:.2f}")
````

````python
name = "Python"

print(f"|{name:<15}|")   # Left

print(f"|{name:^15}|")   # Center

print(f"|{name:>15}|")   # Right
````

````python
number = 123456789

print(f"{number:,}")

pi = 3.1415926535

print(f"{pi:.2f}")

print(f"{pi:.4f}")

percentage = 0.845

print(f"{percentage:.2%}")
````

## Identity Operators
- == -> Compares values.
- is -> Compares identities

````python
# Example 1
a = [10, 20]
b = [10, 20]

print(a == b)
print(a is b)

print("-" * 30)

# Example 2
a = [10, 20]
b = a

print(a == b)
print(a is b)

print("-" * 30)

# Example 3
a = None
b = None

print(a is b)
print(a == b)

print("-" * 30)

# Example 4
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)
print(x == y)

print(x is z)
print(x == z)
````

````python
a = [100, 200]

b = a

c = [100, 200]

print(a == b)
print(a is b)

print(a == c)
print(a is c)

print(id(a))
print(id(b))
print(id(c))
````
