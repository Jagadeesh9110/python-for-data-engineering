# Python Tuples

## Definition

A **tuple** is an **ordered, immutable collection of objects**.

Think of a tuple as a **read-only list**.

Just like a list, a tuple can store:

- Integers
- Floats
- Strings
- Booleans
- Lists
- Dictionaries
- Objects
- Even other tuples

Example:

```python
employee = (101, "Jagadeeswar", True, 55000.0)
```

---

# Why Does Python Provide Tuples?

Python provides tuples because not every collection should be modified.

Some data should remain **constant** after creation.

Examples:

- Employee ID
- GPS Coordinates
- RGB Colors
- Date (Year, Month, Day)
- Database Records
- Function Return Values

Instead of trusting the programmer not to modify the data, Python enforces it by making tuples **immutable**.

---

# Internal Working

A tuple stores **references to its elements**, just like a list.

Example:

```python
numbers = (10, 20, 30)
```

Internally:

```
numbers
    │
    ▼
+----------------------+
| 10 | 20 | 30 |
+----------------------+
```

Each element is actually an object, and the tuple stores references to those objects.

---

# The Main Difference from Lists

Once a tuple is created,

**Python does NOT allow its structure to change.**

This means:

- You cannot add elements.
- You cannot remove elements.
- You cannot replace elements.
- You cannot sort it.
- You cannot reverse it.

The tuple becomes **fixed** after creation.

---

# Why is There No append() Method?

Lists have:

```python
append()
extend()
insert()
remove()
pop()
sort()
reverse()
```

because a list is **mutable**.

A tuple is **immutable**.

Therefore methods like:

```python
append()
remove()
insert()
pop()
sort()
reverse()
clear()
```

do not exist.

Python intentionally does not provide these methods because allowing them would violate the idea of an immutable collection.

# List vs Tuple Internally

## List

✔ Ordered

✔ Mutable

✔ Can Shrink

✔ Can Replace Elements

✔ Supports Many Modification Methods

```
[10, 20, 30]

↓

append()

```

---

## Tuple

✔ Ordered

✔ Immutable

✔ Fixed Size

✔ No Growth

✔ No Shrinking

✔ No Replacement

```
(10, 20, 30)

↓

```

## Single Element Tuple

Incorrect:

```python
t = (10)
```

Output:

```python
<class 'int'>
```

Correct:

```python
t = (10,)
```

Output:

```python
<class 'tuple'>
```

**Important:**

The **comma ( , )** creates the tuple, not the parentheses.

# Packing

Packing means putting multiple values into one tuple.

```python
employee = ("Jagadeeswar",22,"Developer")
```

---

# Unpacking

```python
name, age, role = employee
```

Result

```python
name = "Jagadeeswar"
age = 22
role = "Developer"
```

---

# Extended Unpacking

```python
numbers = (10,20,30,40,50)

first, *middle, last = numbers
```

Result

```python
first = 10
middle = [20,30,40]
last = 50
```

Notice:

Although the source is a tuple,

`middle` becomes a **list**.

The starred (`*`) variable always collects the remaining elements into a list.

---

# Tuple Methods

Unlike lists, tuples have only **two built-in methods**.

## 1. count()

Counts how many times a value appears.

```python
numbers = (10,20,10,30)

numbers.count(10)
```

Output

```python
2
```

---

## 2. index()

Returns the index of the first occurrence.

```python
numbers = (10,20,30)

numbers.index(20)
```

Output

```python
1
```

---

# Methods NOT Available

The following methods do **not exist** for tuples because tuples are immutable.

```python
append()
extend()
insert()
remove()
pop()
clear()
sort()
reverse()
copy()
```

Attempting to use them will raise an `AttributeError`.

Example:

```python
numbers = (10,20,30)

numbers.append(40)
```

Output:

```python
AttributeError: 'tuple' object has no attribute 'append'
```

# Performance

Compared to lists:

- Tuples use **less memory**.
- Tuples are **slightly faster** to create and iterate.
- Tuples are safer because they cannot be modified accidentally.

# Enterprise Use Cases

Tuples are commonly used for:

- Database records
- GPS coordinates
- RGB color values
- Function return values
- Immutable configuration values
- Dictionary keys (because tuples are hashable)

Example:

```python

color = (255,255,255)

employee = (101,"John",50000)
```

# Rule to Remember

**Use a List** when the collection needs to change.

**Use a Tuple** when the collection should remain fixed after creation.

Think of it like this:

- **List = Editable Notebook**
- **Tuple = Printed PDF**

You can edit a notebook whenever you want.

You cannot directly edit a printed PDF—you would need to create a new one.
