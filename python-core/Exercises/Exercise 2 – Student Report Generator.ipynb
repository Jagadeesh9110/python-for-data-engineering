## Problem Statement

A school wants to generate a report card for every student.

Write a function with the following signature:

```python
student_report(name, *marks, **extra)
```

The function should calculate and return a complete report for the student.

---

## Requirements

### Required Parameters

- `name` → Student Name
- `*marks` → Any number of subject marks

---

### Optional Parameters (`**extra`)

The function may receive any of the following optional keyword arguments:

- `bonus_marks`
- `attendance`
- `grade`

None of them are mandatory.

---

## Business Rules

### Total Marks

Calculate the sum of all subject marks.

---

### Bonus Marks

If `bonus_marks` is provided,

add it to the total marks.

If it is not provided,

consider bonus marks as **0**.

---

### Average

Average should be calculated after adding bonus marks.

Formula:

```
Average = Total Marks / Number of Subjects
```

---

### Highest Mark

Find the highest mark.

---

### Lowest Mark

Find the lowest mark.

---

### Pass/Fail

If the average is

- Greater than or equal to **35** → "Pass"
- Otherwise → "Fail"

---

## Return Format

Return a dictionary exactly like this:

```python
{
    "name": "Jagadeeswar",
    "subjects": 5,
    "total": 460,
    "average": 92.0,
    "highest": 98,
    "lowest": 82,
    "result": "Pass"
}
```

## Constraints

- Do NOT use global variables.
- Handle the case where no bonus marks are provided.
- Do NOT print inside the function.
- Return only the dictionary.
- Use meaningful variable names.

````python
def student_report(name:str,*marks:int, **extra) -> dict:
    report=dict()
    report["name"]=name
    report["subjects"]=len(marks)
    total_marks=0
    high,low=marks[0],marks[0]
    for val in marks:
        total_marks+=val
        high=max(high,val)
        low=min(low,val)
    if "bonus_marks" in extra:
        total_marks+=extra["bonus_marks"]
    report["total"]=total_marks
    average=total_marks/len(marks)
    report["average"]=average
    report["highest"]=high
    report["lowest"]=low
    status = "PASS" if average >= 35 else "FAIL"
    report["result"]=status

    return report
    
student1=student_report(
    "Jagadeeswar",
    90,
    95,
    88,
    92,
    85
)
print(f"student 1 report card : {student1}")

student2=student_report(
    "Jagadeeswar",
    90,
    95,
    88,
    92,
    85,
    bonus_marks=10
)
print(f"student 2 report card : {student2}")

student3=student_report(
    "Rahul",
    25,
    30,
    40,
    20,
    attendance=85
)
print(f"student 3 report card : {student3}")
````
