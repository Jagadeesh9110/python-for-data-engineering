## Problem Statement

You are developing a monitoring tool for REST APIs.

Write a function with the following signature:

```python
api_analysis(api_name, *response_times) -> dict
```

The function should analyze the response times of an API.

---

## Parameters

### Required

- `api_name`
- `*response_times`

Example

```python
api_analysis(
    "Login API",
    120,
    95,
    110,
    160,
    135
)
```

---

## Requirements

Calculate

- Total Requests
- Average Response Time
- Fastest Response
- Slowest Response

---

## Return Format

```python
{
    "api_name": "Login API",
    "requests": 5,
    "average_response": 124.0,
    "fastest": 95,
    "slowest": 160
}
```

---

## Constraints

- Do not use global variables.
- Return a dictionary.
- Assume at least one response time is given.

````python
def api_analysis(api_name, *response_times) -> dict:
    # Handle the edge case of an empty arguments list safely
    if not response_times:
        return {
            "api_name": api_name,
            "requests": 0,
            "average_response": 0.0,
            "fastest": None,
            "slowest": None
        }
    
    result={}
    result["api_name"]=api_name
    result["requests"]=len(response_times)
    fast_res=None
    slow_res=None
    total_res=0
    for val in response_times:
        total_res+=val
        if fast_res is None or val < fast_res:
            fast_res=val
        if slow_res is None or val > slow_res:
            slow_res=val

    result["average_response"]=total_res / len(response_times)
    result["fastest"]=fast_res
    result["slowest"]=slow_res

    return result

login_api_analysis= api_analysis(
    "Login API",
    120,
    95,
    110,
    160,
    135
)

print(f"Login api analysis details : {login_api_analysis}")
        
````
