## Problem Statement

You own a coffee shop.

Customers can order multiple cups of coffee and may also choose optional add-ons such as:

- Extra Shot
- Caramel
- Chocolate
- Whipped Cream

Write a function with the following signature:

```python
coffee_bill(cups, *addons, **discount)

```
## Pricing Rules
- Coffee price = ₹120 per cup.
- Every add-on costs ₹25.
- Apply a discount only if discount_percent is provided through **kwargs.
- Return the final bill amount after applying the discount.


````python
def coffee_bill(cups:int,*addons:str, **discount:int) -> float:
    total=0
    total=total + cups*120
    if(len(addons)!=0):
        total+=25*len(addons)

    if "discount_percent" in discount:
        discount_percent = discount["discount_percent"]
        discount_amount = (total * discount_percent) / 100
        total -= discount_amount

    return total


res=coffee_bill(
    2,
    "Caramel",
    "Chocolate",
    discount_percent=10,
    gst=5
)
print(f"Total Cost: {res}")
````
