## Problem Statement

You are developing an HR Management System.

Write a function with the following signature:

```python
employee_salary(employee_name, base_salary, **allowances)
```

The function should calculate the employee's gross salary.

---

## Required Parameters

- employee_name
- base_salary

---

## Optional Allowances

The function may receive any of the following allowances:

- hra
- da
- travel
- bonus
- incentive

All allowances are optional.

---

## Business Rules

- Gross Salary = Base Salary + Sum of all allowances.
- If no allowances are provided, Gross Salary = Base Salary.
- Return the result as a dictionary.

---

## Return Format

```python
{
    "employee": "Jagadeeswar",
    "base_salary": 50000,
    "gross_salary": 68000
}
```

---

## Sample Calls

```python
employee_salary(
    "Jagadeeswar",
    50000
)

employee_salary(
    "Jagadeeswar",
    50000,
    hra=10000,
    da=5000,
    travel=3000
)

employee_salary(
    "Rahul",
    70000,
    hra=12000,
    bonus=10000,
    incentive=5000
)
```

````python
def employee_salary(employee_name, base_salary, **allowances) -> dict:
    result={}
    result["employee"]=employee_name
    result["base_salary"]=base_salary
    gross_salary=base_salary
    if "hra" in allowances:
        gross_salary+=allowances["hra"]
    if "da" in allowances:
        gross_salary+=allowances["da"]
    if "travel" in allowances:
        gross_salary+=allowances["travel"]
    if "bonus" in allowances:
        gross_salary+=allowances["bonus"]
    if "incentive" in allowances:
        gross_salary+=allowances["incentive"]
    result["gross_salary"]=gross_salary

    return result

emp1=employee_salary(
    "Jagadeeswar",
    50000
)

print(f"Employee 1 salary details: {emp1}")

emp2=employee_salary(
    "Jagadeeswar",
    50000,
    hra=10000,
    da=5000,
    travel=3000
)
print(f"Employee 2 salary details: {emp2}")

emp3=employee_salary(
    "Rahul",
    70000,
    hra=12000,
    bonus=10000,
    incentive=5000
)
print(f"Employee 3 salary details: {emp3}")
        
````
