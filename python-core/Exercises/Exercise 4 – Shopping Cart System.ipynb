## Problem Statement

You are building an e-commerce application.

Write a function

```python
shopping_cart(customer_name, **products)
```

Each keyword argument represents a product and its price.

---

## Example

```python
shopping_cart(
    "Jagadeeswar",
    Laptop=65000,
    Mouse=700,
    Keyboard=1500,
    Bag=2000
)
```

---

## Requirements

Calculate

- Total Bill
- Number of Products
- Costliest Product
- Cheapest Product

Return everything as a dictionary.

---

## Expected Return

```python
{
    "customer": "Jagadeeswar",
    "products": 4,
    "total": 69200,
    "costliest": "Laptop",
    "cheapest": "Mouse"
}
```

````python
def shopping_cart(customer_name, **products) -> dict:
    result={}
    result["customer"]=customer_name
    result["products"]=len(products)
    costliest=None
    cheapest=None
    total=0;
    for val in products.values():
        total+=val
    result["total"]=total
    for key,value in products.items():
        if costliest is None or value> costliest:
            costliest=value
            result["costliest"]=key
        if cheapest is None or value< cheapest:
            cheapest=value
            result["cheapest"]=key


    return result

cart1=shopping_cart(
    "Jagadeeswar",
    Laptop=65000,
    Mouse=700,
    Keyboard=1500,
    Bag=2000
)

print(f" Shopping Cart 1 details: {cart1}")
        
````
