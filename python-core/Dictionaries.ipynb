# Python Dictionaries

## Definition

A **dictionary** is an **ordered, mutable collection of key-value pairs**.

Example:

```python
employee = {
    "id":101,
    "name":"Jagadeeswar",
    "salary":50000
}
```

Instead of storing only values,

a dictionary stores:

```

Key

↓

Value

```

---

# Why Does Python Provide Dictionaries?

Suppose we have:

```python
employee = ["Jagadeeswar",22,50000]
```

Question:

What does

```python
employee[2]
```

mean?

Salary?

Age?

No idea.

Instead:

```python
employee = {
    "name":"Jagadeeswar",
    "age":22,
    "salary":50000
}
```

Now:

```python
employee["salary"]
```

Much clearer.

### Accessing Values

````python
employee = {
    "name":"Jagadeeswar",
    "salary":50000
}

print(employee["salary"])# If key doesn't exist then it gives keyError

print(employee.get("age")) # None if noe exist
print(employee.get("age",0)) # Provide default value if not exist
````

### Adding new Data

### Updating Data

### Dictionary Methods

````python
employee = {
    "id": 101,
    "name": "Jagadeeswar",
    "salary": 50000
}

print(employee.keys()) #Returns all the keys in the dictionary.
print(employee.values()) #Returns all the values.

print(employee.items()) #Returns both key and value together.Each item is returned as a tuple.


for key, value in employee.items():
    print(key, value)

employee = {
    "name":"Jagadeeswar"
}
 #Used to merge another dictionary.If a key already exists:python updates the value.
employee.update({
    "salary":50000,
    "city":"Kadapa"
})

print(employee)
````

````python
employee = {
    "name":"Jagadeeswar",
    "salary":50000
}

salary = employee.pop("salary") # Removes the key and returns its value.

print(salary)

print(employee)
````

````python
employee = {
    "name":"Jagadeeswar",
    "salary":50000,
    "city":"Kadapa"
}

print(employee.popitem())#Remvoes the last inserted key-value pair.

employee.clear()# clear the entire dictionary

print(employee)
````

````python
employee = {
    "name":"Jagadeeswar"
}

new_employee = employee.copy() #Creates the another dictionary.his is a shallow copy.Different object.Same contents.
print(new_employee)
print(id(employee)==id(new_employee))
````

````python
employee = {
    "name":"Jagadeeswar"
}
employee.setdefault("salary",50000) # if key doesn't exist then it adds that otherwise nothiung changes.
print(employee)
employee.setdefault("salary",70000)
print(employee)
````

````python
keys = ["name","age","salary"]

employee = dict.fromkeys(keys)#Creates a new dictionary using a collection of keys with values None.

print(employee)
employee = dict.fromkeys(keys,0)#Creates a new dictionary using a collection of keys with a default value.
print(employee)
````

````python
employee = {
    "name":"Jagadeeswar",
    "salary":50000,
    "city":"Kadapa"
}

for key in employee:
    print(key)

for key in employee.keys():
    print(key)

for value in employee.values():
    print(value)

for key, value in employee.items():
    print(key, value)
````

````python
employee = {
    "id":101,
    "name":"Jagadeeswar",
    "address":{
        "city":"Kadapa",
        "state":"Andhra Pradesh"
    }
}

print(employee)
print(employee["address"]["city"]) #Accessing the nested value.This is exactly how JSON from REST APIs is structured.
````
