# File Handling

A file is a collection of data stored permanently on secondary storage such as a hard disk or SSD.

Unlike variables,

which store data temporarily in RAM,

files allow data to persist even after the program terminates.

Examples of files include:

- Text files (.txt)
- CSV files (.csv)
- JSON files (.json)
- Excel files (.xlsx)
- Images (.png, .jpg)
- PDF documents (.pdf)
- Log files (.log)

Python provides built-in support for creating, reading, updating, and deleting files.

Programs store variables in memory (RAM).

When the program ends,

all variables are destroyed.

If we want data to be available after the program finishes,

we must store it in a file.

Examples:

- Employee records
- Student marks
- Configuration files
- Application logs
- Sales reports
- Machine learning datasets
- ETL pipeline input/output files

File Handling allows programs to save and retrieve this persistent data.

| RAM                         | File                             |
| --------------------------- | -------------------------------- |
| Temporary                   | Permanent                        |
| Fast                        | Slower than RAM                  |
| Data lost when program ends | Data persists after program ends |
| Used during execution       | Used for long-term storage       |

# The open() Function

The `open()` function is used to open a file.

It establishes a connection between your Python program and a file stored on the operating system.

Once the file is opened,

Python returns a **File Object**.

All subsequent operations such as reading, writing, and closing the file are performed using this file object.

## Syntax

```python
file = open(file_path, mode="r", encoding="utf-8")
```


| Parameter   | Description                                |
| ----------- | ------------------------------------------ |
| `file_path` | Path of the file                           |
| `mode`      | How the file should be opened              |
| `encoding`  | Character encoding (mainly for text files) |


- open() does not return the file contents.

- It returns a File Object.

A File Object is a Python object that represents an open file.

It acts as an interface between your Python program and the operating system.

Using this object,

Python can

- Read data
- Write data
- Move the file pointer
- Close the file
- Retrieve file information

Without the File Object,

Python cannot interact with the file.

Write mode is used to write data into a file.

Characteristics

- Creates the file if it doesn't exist.
- If the file already exists,

- File pointer starts from the beginning.

````python
file = open("students.txt", "w")

file.write("Alice\n")
file.write("Bob\n")
file.write("Charlie\n")

file.close()
````

`w+` allows both writing and reading.

Characteristics

- Creates file if needed.
- Deletes existing contents.
- Pointer starts at beginning.

````python
file = open("students.txt", "w+")

file.write("Jagadeesh")

file.seek(0)

print(file.read())

file.close()
````

## Why Use seek(0)?

```python
# After writing, the file pointer rests at the very end
```
**Pointer position**:
```text
Jagadeesh
          ↑
```

Because **reading starts from the current pointer position**, running a read operation right now:
```python
file.read()
```
Would return an **empty string** `""`.

Therefore, you must reset it:
```python
file.seek(0)
```
This **moves the pointer back** to the beginning of the file so you can read the content.

- `read()` reads from the **current file pointer position**.

- Suitable for small files.
- Avoid using it for very large files because the entire file is loaded into memory.

````python
file = open("students.txt", "r")

print(file.read())

print(file.read())

file.close()
````

## Why Does the Second read() Return an Empty String?

### First read()
The pointer starts at the **beginning** of the file and reads everything:
```text
↓
Alice
Bob
Charlie
David
Eve
```

### After reading
The file pointer moves all the way to the **End Of File (EOF)**:
```text
Alice
Bob
Charlie
David
Eve
       ↑
      EOF
```

### Second read()
Python attempts to read again starting directly from the current pointer location (**EOF**):
* **Result**: Nothing is left to read.
* **Returns**: `''` (an empty string).

# writelines()

`writelines()` writes multiple strings to a file.

Syntax

```python
file.writelines(iterable)
```

The iterable can be:

- List
- Tuple
- Any iterable containing strings

writelines() writes each string one after another.

It does not automatically add newlines.

Python simply writes exactly what is present in each string.

````python
students = [
    "Alice\n",
    "Bob\n",
    "Charlie\n"
]

file = open("students.txt", "w")

file.writelines(students)

file.close()
````

When you call

```python
file.write(...)
```

Python usually does **not** write the data to the disk immediately.

Instead, it stores the data temporarily in a memory area called the **buffer**.

Later, Python sends the buffered data to the operating system, which writes it to the disk.

Buffering improves performance because writing to memory is much faster than writing to disk.

# flush()

The `flush()` method forces Python to immediately write the buffered data to the file.

Syntax

```python
file.flush()
```

````python
file = open("students.txt", "w")

file.write("Alice")

file.flush()

print("Data flushed to disk.")

file.close()
````

- Without flush(), the data may remain in the memory buffer until the buffer is full or the file is closed.

- flush() ensures that the current buffered data is written immediately.

- Python automatically flushes any remaining buffered data before closing the file.

- So, in most cases, you don't need to call flush() manually.

`flush()` is useful in situations where data must be written immediately.

Examples:

- Log files
- Progress tracking
- Real-time monitoring
- Long-running programs
- Streaming applications

# The `with` Statement (Context Managers)

The `with` statement is used to work with resources such as:

- Files
- Database connections
- Network sockets
- Locks
- Threads

It automatically manages the resource.

For files, it automatically closes the file after use, even if an exception occurs.

When execution leaves the block (whether normally or due to an exception), Python automatically executes:You never need to call it yourself.

````python
with open("students.txt", "r") as file:
    content = file.read()
    print(content)
````

## Multiple Files with with

- Copies the content.

````python
with open("students.txt", "r") as source, open("backup.txt", "w") as destination:
    destination.write(source.read())
````

# DictWriter

`DictWriter` writes dictionaries into CSV files.

Keys become column names.

Values become row values.

````python
import csv

with open("students.csv", "w", newline="") as file:

    fieldnames = ["ID", "Name", "Age"]

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()

    writer.writerow({
        "ID": 1,
        "Name": "Alice",
        "Age": 22
    })

    writer.writerow({
        "ID": 2,
        "Name": "Bob",
        "Age": 23
    })
````

* **Data writing**: Writes a single dictionary as one formatted row.

# JSON (JavaScript Object Notation)

JSON stands for **JavaScript Object Notation**.

It is a lightweight data-interchange format used to exchange data between systems.

Although it originated from JavaScript, JSON is language-independent and is supported by almost every programming language.

Example

```json
{
    "id": 1,
    "name": "Alice",
    "age": 22,
    "city": "Hyderabad"
}
```

JSON is popular because it is:

- Human-readable
- Lightweight
- Easy to parse
- Language-independent
- Supported by almost every API

JSON is widely used in:

- REST APIs
- Web Applications
- Mobile Apps
- Cloud Services
- Configuration Files
- MongoDB Documents

| JSON                         | Python Dictionary                   |
| ---------------------------- | ----------------------------------- |
| Stored as text               | Stored as objects in memory         |
| Keys must be strings         | Keys can be many immutable types    |
| Uses `true`, `false`, `null` | Uses `True`, `False`, `None`        |
| Double quotes are standard   | Single or double quotes are allowed |


### JSON vs. Python Dictionary Data Types

```json
// Example JSON
{
    "name": "Alice",
    "age": 22,
    "active": true
}
```

```python
# Equivalent Python Dictionary
{
    "name": "Alice",
    "age": 22,
    "active": True
}
```

* **Boolean differences**: JSON uses lowercase `true`/`false`, while Python requires capitalized `True`/`False`.
* **String syntax**: JSON strictly demands double quotes (`""`), whereas Python accepts single (`''`) or double quotes.

# Serialization

Serialization is the process of converting a Python object into a format that can be stored or transmitted.

In the `json` module:

```python
json.dumps()
json.dump()
```

perform serialization.

### Why Serialization?

Imagine a standard Python dictionary structure:
```python
student = {
    "name": "Alice",
    "age": 22
}
```

* **The problem**: External systems or machines cannot directly parse or interact with native Python objects.
* **The solution**: Python serializes the dictionary, converting it into a standardized JSON string:
```json
{"name":"Alice","age":22}
```

#### Common use cases for serialized data
* **File storage**: Saving structural configurations directly to local files.
* **Network transfer**: Sending data payloads securely between separate web servers.
* **API communication**: Returning data responses from backend API endpoints.
* **Database storage**: Saving dynamic documents into NoSQL databases.

# Pretty Printing JSON

By default,

```python
json.dump()
```

writes JSON on a single line.

Python provides the `indent` parameter to make JSON more readable.

````python
import json

student = {
    "id": 1,
    "name": "Alice",
    "age": 22,
    "city": "Hyderabad"
}

with open("student.json", "w") as file:
    json.dump(student, file, indent=4)
````

### Why Pretty Printing?

#### Without indent
```json
{"id":1,"name":"Alice","age":22}
```

#### With indent=4
```json
{
    "id": 1,
    "name": "Alice",
    "age": 22
}
```

* **Human readability**: Adding indentation structures compact blocks into a clean visual format that is **much easier for humans to read**.


| Parameter               | Purpose                     |
| ----------------------- | --------------------------- |
| `indent=4`              | Pretty formatting           |
| `sort_keys=True`        | Sort keys alphabetically    |
| `ensure_ascii=False`    | Preserve Unicode characters |
| `separators=(",", ":")` | Compact JSON formatting     |
