# File Handling

A file is a collection of data stored permanently on secondary storage such as a hard disk or SSD.

Unlike variables,

which store data temporarily in RAM,

files allow data to persist even after the program terminates.

Examples of files include:

- Text files (.txt)
- CSV files (.csv)
- JSON files (.json)
- Excel files (.xlsx)
- Images (.png, .jpg)
- PDF documents (.pdf)
- Log files (.log)

Python provides built-in support for creating, reading, updating, and deleting files.

Programs store variables in memory (RAM).

When the program ends,

all variables are destroyed.

If we want data to be available after the program finishes,

we must store it in a file.

Examples:

- Employee records
- Student marks
- Configuration files
- Application logs
- Sales reports
- Machine learning datasets
- ETL pipeline input/output files

File Handling allows programs to save and retrieve this persistent data.

| RAM                         | File                             |
| --------------------------- | -------------------------------- |
| Temporary                   | Permanent                        |
| Fast                        | Slower than RAM                  |
| Data lost when program ends | Data persists after program ends |
| Used during execution       | Used for long-term storage       |

# The open() Function

The `open()` function is used to open a file.

It establishes a connection between your Python program and a file stored on the operating system.

Once the file is opened,

Python returns a **File Object**.

All subsequent operations such as reading, writing, and closing the file are performed using this file object.

## Syntax

```python
file = open(file_path, mode="r", encoding="utf-8")
```


| Parameter   | Description                                |
| ----------- | ------------------------------------------ |
| `file_path` | Path of the file                           |
| `mode`      | How the file should be opened              |
| `encoding`  | Character encoding (mainly for text files) |


- open() does not return the file contents.

- It returns a File Object.

A File Object is a Python object that represents an open file.

It acts as an interface between your Python program and the operating system.

Using this object,

Python can

- Read data
- Write data
- Move the file pointer
- Close the file
- Retrieve file information

Without the File Object,

Python cannot interact with the file.

A File Descriptor is a small integer created by the operating system.

It uniquely identifies an open file.

Python stores this descriptor inside the File Object.

When you call

```python

```

Python uses the stored File Descriptor to communicate with the operating system.

The File Descriptor is managed automatically by Python.

## Why ?
Python asks the operating system to open the file.

the operating system reports an error.

Python converts that error into a `FileNotFoundError` exception.

## With Exception handling

````python
try:
    file = open("abc.txt")

except FileNotFoundError:
    print("File does not exist.")
````

- Always close files after use.
- Use relative paths whenever possible.
- Handle `FileNotFoundError` appropriately.
- Specify an encoding (such as UTF-8) when working with text files to avoid encoding-related issues.
- Prefer the `with` statement for automatic resource management (covered later).

# File Modes

When opening a file, Python needs to know **how you want to use the file**.

This is specified using the **mode** parameter of the `open()` function.

Different modes allow different operations, such as:

- Reading
- Writing
- Appending
- Creating new files
- Reading and writing together
- Binary operations

Syntax:

```python
open(file_path, mode)
```

| Mode | Meaning          | File Must Exist | Creates File | Deletes Existing Data |
| ---- | ---------------- | --------------- | ------------ | --------------------- |
| `r`  | Read             | ✅ Yes           | ❌ No         | ❌ No                  |
| `w`  | Write            | ❌ No            | ✅ Yes        | ✅ Yes                 |
| `a`  | Append           | ❌ No            | ✅ Yes        | ❌ No                  |
| `x`  | Exclusive Create | ❌ No            | ✅ Yes        | ❌ No                  |
| `r+` | Read + Write     | ✅ Yes           | ❌ No         | ❌ No                  |
| `w+` | Read + Write     | ❌ No            | ✅ Yes        | ✅ Yes                 |
| `a+` | Read + Append    | ❌ No            | ✅ Yes        | ❌ No                  |


## Write Mode(w)

Write mode is used to write data into a file.

Characteristics

- Creates the file if it doesn't exist.
- If the file already exists,
  **all existing contents are deleted**.
- File pointer starts from the beginning.

````python
file = open("students.txt", "w")

file.write("Alice\n")
file.write("Bob\n")
file.write("Charlie\n")

file.close()
````

## Read Mode (`r`)

Read mode is the default mode.

Purpose:

- Read the contents of a file.

Characteristics:

- File **must exist**.
- Cannot modify the file.
- If the file doesn't exist, Python raises `FileNotFoundError`.

````python
file = open("students.txt", "r")

content = file.read()

print(content)

file.close()
````

# Read and Write Mode (`r+`)

`r+` allows both reading and writing.

Characteristics

- File must exist.
- Existing contents are preserved.
- Pointer starts at the beginning.

````python
file = open("students.txt", "r+")

print(file.read())

file.write("\nDavid")

file.close()
````

# Write and Read Mode (`w+`)

`w+` allows both writing and reading.

Characteristics

- Creates file if needed.
- Deletes existing contents.
- Pointer starts at beginning.

````python
file = open("students.txt", "w+")

file.write("Jagadeesh")

file.seek(0)

print(file.read())

file.close()
````

## Why Use seek(0)?

```python
# After writing, the file pointer rests at the very end
```
**Pointer position**:
```text
Jagadeesh
          ↑
```

Because **reading starts from the current pointer position**, running a read operation right now:
```python
file.read()
```
Would return an **empty string** `""`.

Therefore, you must reset it:
```python
file.seek(0)
```
This **moves the pointer back** to the beginning of the file so you can read the content.

# Reading Files

Python provides multiple ways to read data from a file.

Each method serves a different purpose depending on:

- File size
- Memory usage
- Performance
- Reading strategy

Common Reading Methods

- read()
- read(size)
- readline()
- readlines()
- for line in file

````python
file = open("students.txt", "w")

file.write("Alice\n")
file.write("Bob\n")
file.write("Charlie\n")
file.write("David\n")
file.write("Eve")

file.close()
````

# read()

The `read()` method reads the entire contents of the file.

Syntax

```python
file.read()
```

Return Type

`str`

````python
file = open("students.txt", "r")

content = file.read()

print(content)

file.close()
````

↓

↓

↓

- `read()` reads from the **current file pointer position**.
- After `read()`, the pointer moves to the end of the file.
- Calling `read()` again without resetting the pointer returns an empty string.
- Suitable for small files.
- Avoid using it for very large files because the entire file is loaded into memory.

````python
file = open("students.txt", "r")

print(file.read())

print(file.read())

file.close()
````

## Why Does the Second read() Return an Empty String?

### First read()
The pointer starts at the **beginning** of the file and reads everything:
```text
↓
Alice
Bob
Charlie
David
Eve
```

### After reading
The file pointer moves all the way to the **End Of File (EOF)**:
```text
Alice
Bob
Charlie
David
Eve
       ↑
      EOF
```

### Second read()
Python attempts to read again starting directly from the current pointer location (**EOF**):
* **Result**: Nothing is left to read.
* **Returns**: `''` (an empty string).

## Looping through readlines()

````python
file = open("students.txt", "r")

students = file.readlines()

for student in students:
    print(student.strip())

file.close()
````

# File Pointer

A file pointer is an internal cursor that keeps track of the current reading or writing position inside a file.

Whenever we perform

- read()
- write()
- readline()

the pointer automatically moves.

Think of it like the cursor while typing in a text editor.

# tell()

`tell()` returns the current position of the file pointer.

Syntax

```python
file.tell()
```

Return Type

Integer

````python
file = open("students.txt", "r")

print(file.tell())

file.close()
````

````python
file = open("students.txt", "r")

file.read(5)

print(file.tell())

file.close()
````

````python
file = open("students.txt", "r")

print(file.read(8))

print(file.tell())

file.close()
````

# seek()

`seek()` moves the file pointer to a specific position.

Syntax

```python
file.seek(position)
```

The next read or write starts from that position.

````python
file = open("students.txt", "r")

file.seek(0)

print(file.read())

file.close()
````

````python
file = open("students.txt", "r")

file.read(5)

print(file.read())

file.seek(0)

print(file.read())

file.close()
````

## seek() to Any Position

````python
file = open("students.txt", "r")

file.seek(6)

print(file.read())

file.close()
````

````python
file = open("students.txt", "r")

print(file.tell())

file.read(10)

print(file.tell())

file.seek(0)

print(file.tell())

file.close()
````

````python
file = open("students.txt", "r")

print(file.read(5))

print(file.tell())

file.seek(0)

print(file.read())

file.close()
````

These methods are useful when you want to:

- Read a specific part of a file.
- Re-read previously read data.
- Skip unwanted data.
- Resume processing from a known position.
- Implement checkpoint-based file processing.

They are commonly used in:

- Log file analysis
- Binary file processing
- Video/audio processing
- Large data pipelines
