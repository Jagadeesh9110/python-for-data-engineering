# Python Sets

## Definition

A **set** is an **unordered, mutable collection of unique objects**.

Unlike a list:

- A set does **not allow duplicate values**.
- A set is **not indexed**.
- A set is optimized for **fast searching**.

Example:

```python
numbers = {10, 20, 30}
```

---

# Why Does Python Provide Sets?

Imagine an employee database.

```python
employees = [101,102,103,101,104,102]
```

There are duplicate IDs.

Removing duplicates using lists is slow.

Instead:

```python
employees = {101,102,103,104}
```

Sets automatically remove duplicates.

---

# Internal Working

Lists are internally implemented using **Dynamic Arrays**.

Sets are internally implemented using a **Hash Table**.

Conceptually:

```

Value

↓

Hash Function

↓

Hash Code

↓

Hash Table

↓

Store Object

```

This allows extremely fast operations like:

- Searching
- Adding
- Removing

Average Time Complexity:

```

Search : O(1)
Insert : O(1)
Delete : O(1)

```

---

# Why Are Duplicates Not Allowed?

When inserting:

```python
numbers = {10,20,30}
```

Python computes a hash.

When adding:

```python
numbers.add(20)
```

Python checks:

```

Hash Exists?

↓

Yes

↓

Ignore

```

So duplicates are automatically removed.

# Rule to Remember

Use a **Set** when:

- Duplicate values are not allowed.
- Fast searching is required.
- Comparing collections.
- Membership checking (`in`).

````python
numbers = {}

print(type(numbers)) # this will show type as dictionary ,Because Python interprets {} as an empty dictionary, not a set.
````

````python
numbers = set()

print(type(numbers))
````

## Creating a Set with Values

````python
numbers = {10, 20, 30}

print(numbers)
print(type(numbers))
````

## Creating a Set from a List

````python
numbers = [10, 20, 20, 30, 30]

unique_numbers = set(numbers)

print(unique_numbers)
````

## Creating from a Tuple

````python
numbers = (10,20,20,30)

unique = set(numbers)

print(unique)
````

## Creating from a String

````python
letters = set("programming")

print(letters)
````

## Creating from a Range

````python
numbers = set(range(1,6))

print(numbers)
````

## Unlike the lists , sets are not accessed through the indexing because sets store thorugh the hash values.

## To Acees the elements we should use the loop.

````python
numbers = {10,20,30}

for number in numbers:
    print(number)
````

## Membership testing

````python
numbers = {10,20,30}

print(20 in numbers)
````

## Adding Elements

````python
numbers = {10,20}

numbers.add(30) # Adds a single element, no duplicates are alllowed

print(numbers)


numbers.update([30,40,50]) # Adds the multiple elements

print(numbers)
````

## Removing Elements

````python
numbers = {10,20,30}

numbers.remove(20) # If element doesn't exist: keyError 

print(numbers)

numbers.discard(100) #No error.This is safer when you're unsure whether the value exists.

numbers = {10,20,30}

value = numbers.pop() # Python removes an arbitrary element.Do not assume it's the first or last element.

print(value)
print(numbers)

numbers = {10,20,30}

new_numbers = numbers.copy() # Creates another set.Just like lists, this is a shallow copy.
print(new_numbers)
print(id(numbers)==id(new_numbers))
````

## Sets Operations


````python
a = {1,2,3}
b = {3,4,5}

# c=a | b #union
c=a.union(b)
print(c)
````

````python
a = {1,2,3}
b = {2,3,4}

print(a & b) #Intersection
````

````python
a = {1,2,3}
b = {2,3,4}

print(a - b) #Difference(elements that exist in the first set.
````

````python
a = {1,2,3}
b = {2,3,4}

print(a ^ b) # Sysmetric Difference. elements that are different.
````
