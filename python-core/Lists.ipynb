# Python List Methods – Quick Reference

## Methods that MODIFY the Original List (In-place)

These methods change the existing list object. They **do not create a new list**.

| Method | Purpose | Returns |
|--------|---------|---------|
| `append(x)` | Adds one element to the end | `None` |
| `extend(iterable)` | Adds multiple elements to the end | `None` |
| `insert(index, x)` | Inserts an element at a specific index | `None` |
| `remove(x)` | Removes the first occurrence of a value | `None` |
| `pop()` / `pop(index)` | Removes and returns an element | Removed element |
| `clear()` | Removes all elements | `None` |
| `sort()` | Sorts the list in ascending order (default) | `None` |
| `reverse()` | Reverses the list | `None` |

---

## Methods that CREATE a New List

These methods return a **new list** without modifying the original list.

| Method | Purpose |
|--------|---------|
| `copy()` | Creates a shallow copy of the list |
| `sorted(list)` | Returns a new sorted list |
| List Comprehension | Creates a new transformed/filtered list |

---

## Methods that ONLY Return Information

These methods do **not** modify the list.

| Method | Returns |
|--------|---------|
| `count(x)` | Number of occurrences of `x` |
| `index(x)` | Index of the first occurrence of `x` |
| `len(list)` | Total number of elements |

---

# Easy Way to Remember

### ✅ Changes the Existing List

append()
extend()
insert()
remove()

clear()

### ✅ Only Gives Information

```python

```

Think:

> **"I am only asking a question about the list."**

# Enterprise Tip

A very common beginner mistake is:

```python
numbers = [3, 1, 2]

numbers = numbers.sort()   # ❌ Wrong
```

Why?

Because:

```python
numbers.sort()
```

modifies the list **in-place** and returns `None`.

Correct:

```python
numbers.sort()

print(numbers)
```

or

```python
sorted_numbers = sorted(numbers)
```

if you want to keep the original list unchanged.

---

# Rule of Thumb

If a method **changes the list**, it usually returns **None**.

If you need a **new list**, use:

- `copy()`
- `sorted()`
- List Comprehensions

instead of modifying the original list.
